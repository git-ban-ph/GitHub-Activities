```python
fruits = ['apple', 'banana', 'grape']
```

# Adding Elements

```python
fruits = ['apple', 'banana', 'grape']
```

```python
fruits.append('orange')
fruits
```

```python
fruits.insert(1, 'blueberry')
fruits
```
