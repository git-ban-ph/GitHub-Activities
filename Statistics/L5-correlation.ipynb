```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Current Power

```python
df = pd.read_csv(r'Datasets\current-power.csv', delimiter=",")
df
```

```python
df.describe()
```

```python
df.cov()
```

```python
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x, y)
plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.show()
```

```python
df.corr()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.Figure(figsize=(7, 4))
plt.scatter(x, y,
            color='#000000',
            alpha=0.6,
            label='Data points'
            )
plt.plot(x, line,
         color="#005555",
         label=f'Regression line (r = {r_value:.3f})'
         )
plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel("Ampere (A)")
plt.legend()
plt.show()
```

## Air Quality <hr style="border: 2.5px solid; color: #d020b0ff"></hr>

```python
air = pd.read_csv(r'Datasets\air-quality.csv', delimiter=",")
air
```

```python
x = air['NOx']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.Figure(figsize=(7, 4))
plt.scatter(x, y,
            color="#3B3B3B",
            alpha=0.5,
            label='Data points'
            )
plt.plot(x, line,
         color="#BFA600",
         label=f'Regression line (r = {r_value:.3f})'
         )
plt.title('Nox vs Humidity Comparison')
plt.ylabel('Relative Humidity')
plt.xlabel("NOx (ug/cu.m)")
plt.legend()
plt.show()
```

```python
x = air['NO2']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.Figure(figsize=(7, 4))
plt.scatter(x, y,
            color="#3B3B3B",
            alpha=0.5,
            label='Data points'
            )
plt.plot(x, line,
         color="#BFA600",
         label=f'Regression line (r = {r_value:.3f})'
         )
plt.title('No2 vs Humidity Comparison')
plt.ylabel('Relative Humidity')
plt.xlabel("NO2 (ug/cu.m)")
plt.legend()
plt.show()
```
