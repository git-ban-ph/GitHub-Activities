```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'dataset\U.S._crude_oil_production_edit.csv', delimiter=',')
```

```python
df.head()
```

```python
# storage of means
means_data = {col: [] for col in df.columns}
```

```python
np.random.seed(42)
numeric_df = df.select_dtypes(include=np.number)

# sampling means
for i in range(50): # 50 times
    for c in df.columns:
        sample = np.random.choice(numeric_df[c], size=30, replace=True)
        means_data[c].append(np.mean(sample))
```

```python
df_means = pd.DataFrame(means_data)
df_means
```

```python
df_means.to_csv(r'dataset\dataset.csv', index=False)
```
